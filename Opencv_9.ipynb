```python
import cv2 as cv
import numpy as np
```

```python
img=cv.imread('image3.jpg')
#img=cv.resize(img, (0,0), fx=0.2, fy=0.2) 

ret,treshold=cv.threshold(img,10,255,cv.THRESH_BINARY)

cv.imshow('image1',img)
cv.imshow('Binary',treshold)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img=cv.imread('image3.jpg')
#img=cv.resize(img, (0,0), fx=0.2, fy=0.2) 
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
ret,treshold=cv.threshold(gray,10,255,cv.THRESH_BINARY)

cv.imshow('image1',img)
cv.imshow('Binary',treshold)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
th = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 115, 1)
cv.imshow('original',img)
cv.imshow('Adaptive threshold',th)
cv.waitKey(0)
cv.destroyAllWindows()
```
