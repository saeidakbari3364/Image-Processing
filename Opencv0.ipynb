# Different Chanelles of  RGB Image 

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img=cv.imread("lenna.png")
```

```python
cv.imshow("image",img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img_RGB=cv.cvtColor(img,cv.COLOR_BGR2RGB)
```

```python
R=img_RGB[...,0]
G=img_RGB[...,1]
B=img_RGB[...,2]
```

```python
 #[rows, columns, channels]
```

```python
print(img_RGB.shape)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1,5,1)
plt.title("BGR Image")
plt.imshow(img)
plt.axis(False)

plt.subplot(1,5,2)
plt.title("RGB Image")
plt.imshow(img_RGB)
plt.axis(False)

plt.subplot(1,5,3)
plt.title("R Channel")
plt.imshow(R)
plt.axis(False)

plt.subplot(1,5,4)
plt.title("G Channel")
plt.imshow(G)
plt.axis(False)

plt.subplot(1,5,5)
plt.title("B Channel")
plt.imshow(B)
plt.axis(False)



plt.show()
```
