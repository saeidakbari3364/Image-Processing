```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
#laod image
image=cv.imread('/content/roadnorway.jpg',0)
```

```python
image_noise_remove=cv.GaussianBlur(image,(3,3),0)
#Laplacian
laplacian=cv.Laplacian(image_noise_remove,cv.CV_64F)
#Sobelx
sobelx=cv.Sobel(image_noise_remove,cv.CV_64F,1,0,ksize=5)
#sobely
sobely=cv.Sobel(image_noise_remove,cv.CV_64F,0,1,ksize=5)

#canny Edge Detection

canny=cv.Canny(image_noise_remove,100,200)
```

```python
#show image
%matplotlib inline
plt.subplot(2,3,1)
plt.title('Original')
plt.imshow(image,cmap='gray')

plt.subplot(2,3,2)
plt.title('laplacian')
plt.imshow(laplacian,cmap='gray')

plt.subplot(2,3,3)
plt.title('Sobelx')
plt.imshow(sobelx,cmap='gray')

plt.subplot(2,3,4)
plt.title('Sobely')
plt.imshow(sobely,cmap='gray')

plt.subplot(2,3,5)
plt.title('canny')
plt.imshow(canny,cmap='gray')

plt.show()
```
