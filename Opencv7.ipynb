# Corner Detection

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('chessboard.jpg')
#resize
img=cv2.resize(img,(0,0),fx=0.5,fy=0.5)
#rgb to gray
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)


#Corners Detection

#100 number of corners-0.01 minimum quality of corner betwwn 0 and 1- 10 minimum eculidian distance between corners
corners=cv2.goodFeaturesToTrack(gray,100, 0.01, 10)

#print(corners)

#we need to convert float corners to integer
corners=np.int0(corners)

```

```python
#Draw these corners

for corner in corners:
    x,y=corner.ravel()
    print(x,':',y)
    cv2.circle(img,(x,y),5,(255,0,0),-1)
    
print(len(corners))
#draw line from corners
for i in range(len(corners)):
    for j in range(i+1,len(corners)):
               print("i: ",i)
               print("j: ",j) 
               c1=tuple(corners[i][0])
               print("c1: ",c1) 
               c2=tuple(corners[j][0])
               print("c2: ",c2) 
               color=tuple(map(lambda x:int(x),np.random.randint(0,255 , size=3)))
               cv2.line(img,c1,c2,color,1)
               print("----") 
               
```

```python
plt.imshow(img)
cv2.imshow('Chess Board',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
